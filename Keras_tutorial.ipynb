## Import relevant packages

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers
import time
import visualkeras
```

## Create a NN using keras

1. We are creating a NN that takes a single scalar-valued input
2. Here, the NN has been defined in such a way that the input layer is explicity initialized. In more recent versions of Keras, this is not required and aligns more closely with other NN frameworks like PyTorch
3. We build here a NN with a single scalar-valued input represented by a single neuron in the input layer, followed by two fully connected hidden layers initialized by the Dense() command having 8 neurons each and the hyperbolic tangent activation function. Finally, an ouput layer with just a single neuron has been used, as we are solving a single-variable regression problem
4. We next initialize a model in keras based on the input and the output
5. We then use the command "model.summary()" to print out information summarizing the architecture of our NN
6. Note that the design choice of our NN here is arbitrary and our goal in this tutorial is familiarize ourselves with the basic computational workflow in Keras

```python
input = keras.Input(shape=(1,))
hidden_1 = layers.Dense(8, activation="tanh")(input)
hidden_2 = layers.Dense(8, activation="tanh")(hidden_1)
output = layers.Dense(1)(hidden_2)
model = keras.Model(input, output)
model.summary()
```

## Visualize NN

```python
visualkeras.layered_view(model)
```

## Initialize the optimizer

1. We will use the popular ADAM optimizer
2. Note that we use a learning rate of 0.03 while the default learning rate is 0.01

```python
opt = keras.optimizers.Adam(learning_rate=0.03, beta_1=0.9, beta_2=0.999)
```

## Configure the model for training

"Configure" here simply means "Setting up"

```python
model.compile(optimizer=opt,loss=tf.keras.losses.MeanSquaredError())
```

## Create training data for regression problem

```python
x = tf.linspace(0.0,4.0*np.pi,200)
y = tf.sin(x)
```

## Fit data

```python
model.fit(x,y,epochs=1500)
```

## Generate prediction

```python
y_pred = model(x)
```

## Plot results

```python
plt.figure()
plt.plot(x,y,'b-',label='Exact, y=sin(x)')
plt.plot(x,y_pred[:,0],'r--', label='Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Single-variable regression using NN in Keras")
plt.legend()
plt.grid()
plt.show()
```

## Automatic differentiation

1. We first set up an input tensor using tf.linspace which houses an array of uniformly spaced out points
2. Next, we convert the input tensor into a TensorFlow variable. This is required to demonstrate automatic differentiation using GradientTape
3. GradientTape is a tape that records the operations required to compute gradients. In cases where we wish to compute multiple gradients, we need to set the "persistent" flag to "True" to inform TensorFlow to not release the GradientTape after computing the first gradient 
4. The watch command ensured that the non-trainable variables with respect to which we wish to compute gradients are recorded on this "tape"
5. The gradient line computes the gradient

### Single input and Single output: First Derivative

```python
xs1 = tf.linspace(0,10,50)
xs1 = tf.Variable(xs1)
with tf.GradientTape() as t:
    t.watch(xs1)
    ys1 = xs1*xs1
ys1_g1 = t.gradient(ys1, xs1)
```

### Plot results

```python
plt.figure()
plt.plot(xs1, ys1, label="function y(x) = x^2")
plt.plot(xs1, ys1_g1, label="first derivative y'(x) = 2x")
plt.xlabel('x')
plt.title('first derivatives in case of single input and single output')
plt.grid()
plt.legend()
plt.show()
```

### Single input and Multiple outputs: First derivative

```python
xs2 = tf.linspace(0,10,50)
xs2 = tf.Variable(xs2)
with tf.GradientTape(persistent=True) as t:
    ym1_i1 = xs2*xs2
    ym1_i2 = xs2*xs2*xs2
ym1_i1_g1 = t.gradient(ym1_i1,xs2)
ym1_i2_g1 = t.gradient(ym1_i2,xs2)
```

### Plot results

```python
plt.figure()
plt.plot(xs2,ym1_i1, label="y(x) = x^2")
plt.plot(xs2,ym1_i2, label="y(x) = x^3")
plt.xlabel('x')
plt.title('functions in case of single input and multiple output')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(xs2,ym1_i1_g1, label="y(x) = x^2 , y'(x) = 2x")
plt.plot(xs2,ym1_i2_g1, label="y(x) = x^3 , y'(x) = 3(x^2)")
plt.xlabel('x')
plt.title('first derivative in case of single input and multiple output')
plt.grid()
plt.legend()
plt.show()
```

### Single input and Single output: Second and First Derivatives

```python
xs3 = tf.linspace(0,10,50)
xs3 = tf.Variable(xs3)
with tf.GradientTape() as t:
    with tf.GradientTape() as t1:
        ys2 = xs3*xs3
    ys2_g1 = t1.gradient(ys2,xs3)
ys2_g2 = t.gradient(ys2_g1,xs3)
```

### Plot results

```python
plt.figure()
plt.plot(xs3, ys2, label="function y(x) = x^2")
plt.plot(xs3, ys2_g1, label="first derivative y'(x) = 2x")
plt.plot(xs3, ys2_g2, label="second derivative y''(x) = 2")
plt.xlabel('x')
plt.title('second and first derivatives in case of single input and single output')
plt.grid()
plt.legend()
plt.show()
```
